## FILTERING DATA 

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 60)]
new_df
```

```python
##---------------------Store data into new datasets 
new_df.to_csv('Filter.csv')
##---------------------Problem is index are same as previous one so reset index
new_df = new_df.reset_index()
```
